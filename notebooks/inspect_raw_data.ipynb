### Inspect Raw Data


```python
import pandas as pd 
from loguru import logger

def read_raw_data():
    """Read the raw data from a CSV file: data/raw/loan.csv"""
    
    FILE_PATH = '../data/raw/loan.csv'
    try:
        df = pd.read_csv(FILE_PATH)
        logger.success(f"Raw data read successfully with {len(df)} rows and {len(df.columns)} columns.")
        return df
 
    except FileNotFoundError:
        logger.error(f"File not found at path {FILE_PATH}. Please check if file exist at the specified path.")
        return None
     
    except Exception as e: 
        logger.error(f"Error reading raw data: {e}")
        return None
```
